# Bayesian Hierarchical Modeling in PyMC3

- https://towardsdatascience.com/bayesian-hierarchical-modeling-in-pymc3-d113c97f5149

The general idea here is that if you have large categories in your model that may differentiate the target quite a bit, you can build sub-models instead of one large model. The reason it's hierachical instead of mutliple single models is that the features/parameters/weights/coefficient share information, which makes the model that much more robust.

## Imports

```python
import numpy as np
import pymc3 as pm
import arviz as az

np.random.seed(0) # to keep it reproducible
```

## Create dataset...

```python
# average slope of 8 groups is 2

mean_slope = 2 # the 8 different slopes have a mean of 2

# create the individual slopes
slopes = np.random.normal(mean_slope, size=8)

# Create the groups each observation belongs to (50 observations for each group, then a minotirty group at the end with only 5 observations)
groups = np.array(50*[0, 1, 2, 3, 4, 5, 6] + 5*[7])

# Just a randon variable
x = np.random.randn(355)

# 
y = slopes[groups] * x + 0.1*np.random.randn(355)
```

```python
# Add some outliers to the minority group

y[-1] = 30 # the last 5 observations are from minority group 7
y[-2] = 30
```

## Pooled Model

```python
    import matplotlib.pyplot as plt
    import numpy as np
    import scipy.stats as st
    plt.style.use('seaborn-darkgrid')
    x = np.linspace(-5, 5, 1000)
    mus = [0., 0., 0., -2.]
    sigmas = [0.4, 1., 2., 0.4]
    for mu, sigma in zip(mus, sigmas):
        pdf = st.norm.pdf(x, mu, sigma)
        plt.plot(x, pdf, label=r'$\mu$ = {}, $\sigma$ = {}'.format(mu, sigma))
    plt.xlabel('x', fontsize=12)
    plt.ylabel('f(x)', fontsize=12)
    plt.legend(loc=1)
    plt.show()
```

```python
with pm.Model() as pooled_model:
    
    # 3. Our slope is normally distributed, our sd is exponential so it only takes into account positive values
    slope = pm.Normal('slope', 0, 20)
    noise = pm.Exponential('noise', 0.1)
    
    # 2. Next we create a normal distribution that needs a mean & standard deviation (add observed so we know it's not a prior)
    obs = pm.Normal('obs', slope*x, noise, observed=y)
    
    # 1. Start with creating the trace whcih samples from the posterior distribution
    pooled_trace = pm.sample(return_inferencedata=True)
```

```python
az.plot_posterior(pooled_trace, var_names=['slope'])
```

## Unpooled Model

```python
with pm.Model() as unpooled_model:
    
    # We add the shape into the slope as well - slopes should be around 0...with a large standard deviation of 20
    slope = pm.Normal('slope', 0, 20, shape=8)
    noise = pm.Exponential('noise', 10)
    
    # We now add the groups to multiply by the slope
    obs = pm.Normal('obs', slope[groups]*x, noise, observed=y)
    
    # Same trace as above...
    unpooled_trace = pm.sample(return_inferencedata=True)
```

```python
az.plot_posterior(unpooled_trace, var_names=['slope'])
```

All trained independently!

Plot the credible intervals for each model (94% hdi) in a single graph

```python
az.plot_forest(unpooled_trace, var_names=['slope'], combined=True)
```

Wait....why don't our models have slopes that are around 2, like we told it? It's because we introduced outliers to the smallest group which pull it all off

## Partially Pooled aka Hierarchical Model

```python
with pm.Model() as hierarchical_model:
    
    # These are executed first, before our actual priors
    mu_slope = pm.Normal('mu_slope', 0, 1) # hyperprior 1
    sigma_slope = pm.Exponential('sigma_slope', 13) # hyperprior 2
    
    
    slope = pm.Normal('slope', mu_slope, sigma_slope, shape=8) # The 0 mean above is replaced by this mu_slope, and 20 for sd replaced by sigma_slop
    noise = pm.Exponential('noise', 10)
    
    
    obs = pm.Normal('obs', slope[groups]*x, noise, observed=y)
    
    # 0.995 - improve sampling, from nested posterior this is not easy
    hierarchical_trace = pm.sample(
        return_inferencedata=True,
        target_accept=0.995
    )
```

```python
az.plot_posterior(hierarchical_trace)
```

```python
az.plot_forest(hierarchical_trace, var_names=['slope'], combined=True)
```
