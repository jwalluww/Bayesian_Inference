```python
import numpy as np
from scipy.stats import ttest_ind
import pymc3 as pm
import arviz as az


np.random.seed(0)
```

Set up two version, the number of customers for eaach version, and the (cheating here) conversion rate

```python
blue_conversions = np.random.binomial(1, 0.02354199, size=13568)
red_conversions = np.random.binomial(1, 0.02493098, size=17921)

print(f'Blue: {blue_conversions.mean():.3%}')
print(f'Red: {red_conversions.mean():.3%}')
```

# Frequentist Test

Null Hypothesis: Default Control Blue is better and will yield more conversions

```python
print(f'p-value: {ttest_ind(blue_conversions, red_conversions, equal_var=False, alternative="less").pvalue:.1%}')
```

Reject the null hypothesis! Evidence suggests that blue is not better than red

# Bayesian Test

- Why?
    - make statements about one version being better than another with a certain probability
    - no statistical tests - just one generative model & then do bayesian inference

```python
with pm.Model():
    
    # 3 - We grab beta distributions for the uncertainty, this is what we put into the model and get out of the model after the likelihood is integrated in
        # 1,99 if graphed gives you a small converion rate- we are pretty confidnece in this but this won't be too overpowering
    blue_rate = pm.Beta('blue_rate', 1, 99)
    red_rate = pm.Beta('red_rate', 1, 99)
    
    
    # 2 - The likelihood of the model - We use a bernoulli distribution because it fits our binary data
        # Notice the inputs here - it requires observations (our data) and a "p", which is a prior probability. We will use beta because it's made for this case.
    blue_obs = pm.Bernoulli('blue_obs', blue_rate, observed=blue_conversions)
    red_obs = pm.Bernoulli('red_obs', red_rate, observed=red_conversions)
    
    
    # 1 - this is your trace and samples from your posterior distribution
    trace = pm.sample(return_inferencedata=True)
```

```python
az.plot_posterior(trace)
```

- Mean listed above
- 94% probability that the true rate lies somewhere in this range

```python
blue_rate_samples = trace.posterior['blue_rate'].values
red_rate_samples = trace.posterior['red_rate'].values
print(f'Probability that red is better: {(red_rate_samples > blue_rate_samples).mean():.1%}.')
# output (for me): Probability that red is better: 91.7%.
```

We sampled from both posterior distributions in teh trace and stored them in that object. We are simply looking at the percent of observations where read is better than blue.
