## Import Libraries & Data

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, TimeSeriesSplit
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Garve/datasets/4576d323bf2b66c906d5130d686245ad205505cf/mmm.csv',
                parse_dates=['Date'],
                 index_col='Date'
                )
df.head()
```

- How much of the 15,904.11 € sales in the week ending on 2021–10–10 (see table above) was generated by TV advertising? And how much by radio and web banners? And what is the baseline, i.e. the number of sales we would have had without any advertising?

- compute ROIs and optimize spending

- restrict ourselves to additive models

```python
# Split X & Y
X = df.drop(columns=['Sales'])
y = df['Sales']
```

<br>

***

<br>

## Run Baseline Model

Use time series split over k-fold cross validation

```python
lr = LinearRegression()
print(cross_val_score(lr, X, y, cv=TimeSeriesSplit()))
```

Coefficients & intercept

```python
lr.fit(X, y) # refit the model with the complete dataset
print('Coefficients:', lr.coef_)
print('Intercept:', lr.intercept_)
```

Predictions for a single week

```python
print(lr.predict(([[11543.58, 4615.35, 2518.88]])))
```
